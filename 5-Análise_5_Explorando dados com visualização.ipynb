<b>Exploring Data with Visualization</b><br>

<b>Exploring data using visualization features such as histograms and scatter plots.</b> <br>

Questions analyzed:<br>
-Comparison of greenhouse gas emission rating distribution in 2008 and 2018;<br>
-Combined mpg distribution from 2008 to 2018;<br>
-Correlation between displacement (engine displacement) and combined mpg;<br>
-Correlation between greenhouse gas emission rating and combined mpg.<br>


```python
# carregue os conjuntos de dados
import pandas as pd
%matplotlib inline

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')
```

```python
#Comparação da distribuição de classificação de emissão de gases do efeito estufa em 2008 e 2018.
#Comparison of greenhouse gas emission rating distribution in 2008 and 2018.

df_08['greenhouse_gas_score'].plot(kind='hist', title='Greenhouse Gas 08');
```

```python
df_18['greenhouse_gas_score'].plot(kind='hist', title='Greenhouse Gas 18');
```

A distribuição de classificação de emissão de gases do efeito estufa mudou de 2008 para 2018, tornando-se mais assimétrica para a esquerda.<br>

The greenhouse gas emission ranking distribution changed from 2008 to 2018, becoming more asymmetric to the left.

```python
#Distribuição de mpg combinado de 2008 para 2018.
#Combined mpg distribution from 2008 to 2018.

df_08['cmb_mpg'].plot(kind='hist', title='MPG Comb 08');
```

```python
df_18['cmb_mpg'].plot(kind='hist', title='MPG Comb 18');
```

A distribuição de mpg combinado alterou-se de 2008 para 2018 ficando mais assimétrica para a direita, indicando a ocorrência de altos valores com baixa frequência.<br>

The combined mpg distribution changed from 2008 to 2018 becoming more asymmetric to the right, indicating the occurrence of high values with low frequency.

```python
#Correlação entre deslocamento (cilindrada) e mpg combinado.
#Correlation between displacement (engine displacement) and combined mpg.

df_08.plot(kind='scatter', x='cyl', y='cmb_mpg', title='Cyl x Mpg Cmb 08');
```

```python
df_18.plot(kind='scatter', x='cyl', y='cmb_mpg', title='Cyl x Mpg Cmb 18');
```

Deslocamento (cilindrada) cyl e Mpg combinado apresentam uma correlaçao negativa, mostrando uma concentração dos pontos em tendência decrescente, ou seja, conforme a variável independente (Mpg comb) aumenta, a variável dependente diminui (cyl).<br>

Displacement (cylinder capacity) cyl and Mpg combined present a negative correlation, showing a concentration of points in a decreasing tendency, that is, as the independent variable (Mpg comb) increases, the dependent variable decreases (cyl).

```python
#Correlação entre classificação de emissão de gases do efeito estufa e mpg combinado.
#Correlation between greenhouse gas emission rating and combined mpg.

df_08.plot(kind='scatter', x='greenhouse_gas_score', y='cmb_mpg', title='Greenhouse x Mpg Cmb 08');
```

```python
df_18.plot(kind='scatter', x='greenhouse_gas_score', y='cmb_mpg', title='Greenhouse x Mpg Cmb 18');
```

Emissão de gases do efeito estufa e mpg combinado apresentam uma correlaçao positiva, mostrando uma concentração dos pontos em tendência crescente, ou seja, conforme a variável independente (Mpg comb) aumenta, a variável dependente também aumente (Greenhouse gas score).<br>

Greenhouse gas emissions and combined mpg show a positive correlation, showing a concentration of points in an increasing tendency, that is, as the independent variable (Mpg comb) increases, the dependent variable also increases (Greenhouse gas score).
