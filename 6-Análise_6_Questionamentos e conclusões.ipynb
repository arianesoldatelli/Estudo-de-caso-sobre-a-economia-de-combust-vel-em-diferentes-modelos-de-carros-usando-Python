<b>Analisando os tipos de combustíveis (Analyzing fuel types)</b><br> 

O GNC - Gás Natural Comprimido é extraído de matérias orgânicas, compostagens e bagaços, este modelo é mais seguro quando comparado ao GNV, por utilizar as matérias orgânicas o seu gás é mais leve e é evaporado rapidamente pela atmosfera o que evita riscos de incêndio e explosão.<br> 
Fonte: https://transac.com.br/noticia/4/transporte-de-combustiveis/novidades-em-combustivel-conheca-o-gnc

CNG - Compressed Natural Gas is extracted from organic materials, compost and bagasse. This model is safer when compared to VNG, as it uses organic materials, its gas is lighter and evaporates quickly into the atmosphere, which prevents the risk of fire and explosion.<br>
Source:https://transac.com.br/noticia/4/transporte-de-combustiveis/novidades-em-combustivel-conheca-o-gnc

Tido como o combustível mais limpo do mundo, o etanol pode reduzir as emissões de gás carbônico de ponta a ponta (da produção até o uso nos veículos) em até 90% em relação à gasolina e, por ser feito de cana-de-açúcar ou de milho, é renovável, diferentemente do diesel e da gasolina.
Fonte: https://exame.com/negocios/vantajoso-abastecer-etanol/

Considered the cleanest fuel in the world, ethanol can reduce carbon dioxide emissions from end to end (from production to use in vehicles) by up to 90% compared to gasoline and, as it is made from sugarcane or corn and it is renewable, unlike diesel and gasoline.
Source: https://exame.com/negocios/vantajoso-abastecer-etanol/

```python
def calc_fuel_prop(df):
    
    fuels = df['fuel'].unique()
    for fuel in fuels:
        total = df.shape[0]
        print('Combustível: {0} Quantidade: {1}'.format(
                fuel, df.query("fuel == @fuel")['fuel'].count()/total))

print("08")
calc_fuel_prop(df_08)
print("18")
calc_fuel_prop(df_18)
```
