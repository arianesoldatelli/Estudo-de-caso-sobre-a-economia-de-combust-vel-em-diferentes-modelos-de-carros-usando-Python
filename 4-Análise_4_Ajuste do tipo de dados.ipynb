As colunas mpg e os nível de gases de efeito estufa apresentam o mesmo problema (número/número), e foram salvos como strings.<br> 
Veículos com mais de um tipo de combustível (híbridos), terão uma string que armazena dois valores - um para cada. <br> 

The mpg and greenhouse gas levels columns have the same problem (number/number), and were saved as strings.<br>
Vehicles with more than one fuel type (hybrids) will have a string that holds two values - one for each.<br>

```python
# Primeiro, obter todos os híbridos de 2008
# First, get all 2008 hybrids
hb_08 = df_08[df_08['fuel'].str.contains('/')]
hb_08
```

```python
# híbridos em 2018
# hybrids in 2018
hb_18 = df_18[df_18['fuel'].str.contains('/')]
hb_18
```

Para trabalhar com os dois valores separados, cada linha que contém um carro híbrido será dividida em duas novas linhas em dois data frames - uma com os valores do primeiro tipo de combustível (antes da "/") e outra com os valores do segundo tipo de combustível (depois da "/").<br> 

To work with the two separate values, each row containing a hybrid car will be split into two new rows in two data frames - one with the values of the first fuel type (before the "/") and another with the values of the second fuel type (after the "/").<br>

```python
# Primeiro: criar cópias do dataframe com híbridos de 2008
# Forst: make copies of the dataframe with 2008 hybrids

df1 = hb_08.copy()  # data on first fuel type of each hybrid vehicle
df2 = hb_08.copy()  # data on second fuel type of each hybrid vehicle
```

```python
# colunas que devem ser divididas por "/"
# columns that should be divided by "/"

split_columns = ['fuel', 'air_pollution_score', 'city_mpg', 'hwy_mpg', 'cmb_mpg', 'greenhouse_gas_score']

# divisão dos valores usando a função split para cada coluna de cada cópia do dataframe
# splitting the values using the split function for each column of each copy of the dataframe

for c in split_columns:
    df1[c] = df1[c].apply(lambda x: x.split("/")[0])
    df2[c] = df2[c].apply(lambda x: x.split("/")[1])
```

```python
# dataframe que armazena informações sobre o PRIMEIRO tipo de combustível do híbrido (valores antes de "/")
# dataframe that stores information about the FIRST fuel type of the hybrid (values before "/")

df1
```

```python
# dataframe que armazena informações sobre o SEGUNDO tipo de combustível do híbrido (valores depois de "/")
# dataframe that stores information about the SECOND fuel type of the hybrid (values AFTER "/")

df2
```

```python
# combinando as informações para adicionar ao dataframe original (com linhas separadas para cada tipo de combustível de cada veículo)
# combining the information to add to the original dataframe (with separate lines for each fuel type for each vehicle)

new_rows = df1.append(df2)
new_rows
```

```python
# descartar as linhas originais que continham híbridos
# discard the original lines that contained hybrids

df_08.drop(hb_08.index, inplace=True)

# adicionar as novas linhas separadas
# add the new separate lines

df_08 = df_08.append(new_rows, ignore_index=True)
```

```python
# conferir se todas linhas originais com híbridos, que continham "/" foram removidas
# check if all original lines with hybrids, which contained "/" were removed

df_08[df_08['fuel'].str.contains('/')]
```
